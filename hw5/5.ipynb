```python
def div (b, a):
    q = floor(b/a)
    r = b - floor(b/a)*a
    c = q*a + r
    if b == c:  print(str(b)+ " = " + str(q)+ 
                          "\\cdot "+ str(a) +" + "+ str(r) + "\\\\") 
    else: return "error"
```

```python
def divisors_less_than_root_x (x):
    return [y for y in range(2, ceiling(sqrt(x))) if x%y == 0 ]
```

```python
def isPrime (x):
    return divisors_less_than_root_x(x) == []
```

```python
def primes (n):
    N = list(range(2,n+1))
    for x in N:
        for y in N:
            if not(y == x) and (y%x == 0):
                N.remove(y)
    return N
```

```python
def L(q,p):
    if 0 == q%p:
        return -1
    elif not([] == [x for x in r(p) if x == q%p]):
        return 1
    else:
        return -1
```

```python
def phi (n):
    F = factorint(n)
    y = 1
    for x in F:
        y *= (1-1/x)
    return round(n*y)
```

always round when unsure.
