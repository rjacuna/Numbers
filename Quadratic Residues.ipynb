```python
def t(m):
    return [(x,x**2%m) for x in range(1,m)]
def r(m):
    return {x**2%m for x in range(1,m)}
```

```python
r(3)
```

```python
t(19)
```

```python
r(19)
```

```python
def L(q,p):
    if 0 == q%p:
        return 0
    elif not([] == [x for x in r(p) if x == q%p]):
        return 1
    else:
        return -1
```

```python
r(5)
```

```python
[(y,L(y,5)) for y in range(100)]
```

```python
import numpy as np
import matplotlib.pyplot as plt

R = range(50)
vals = np.array([[y,L(y,16)] for y in R]).T

plt.plot(* vals)
plt.scatter(* vals)
plt.show()
```
